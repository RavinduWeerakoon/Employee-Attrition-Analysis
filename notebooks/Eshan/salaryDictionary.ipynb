```python
#  load data set
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_salary = pd.read_csv('../Dataset/salary.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Check if 'Net Salary' has any missing values
print(df_salary['Net Salary'].isnull().sum())
```

```python
# create data frame only include 'Employee_No','Net Salary','Year','month'
df_salary_summery = df_salary[['Employee_No', 'Net Salary', 'year', 'month','Total Working Days']]
```

```python
# Group by 'year' and 'month' and calculate the sum of 'Net Salary'
df_salary_summery_2 = df_salary_summery.groupby(['year', 'month']).sum()
```

```python
df_salary_summery_2.head(25)
```

```python
# i want to get group by Employee_No and i want to get sum of net salaries per year
df_salary_summery_3 = df_salary_summery.groupby(['Employee_No', 'year']).sum()
df_salary_summery_3.head(25)
```

```python
# calculate annual salary
annual_salary = df_salary_summery.groupby(['Employee_No', 'year'])['Net Salary'].sum().reset_index()

# Print the result
annual_salary.head()
```
